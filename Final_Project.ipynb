```python
class Item():
  def __init__(self, name, weight, value):
    self.name = name
    self.weight = weight
    self.value = value
  def get_name(self):
    return self.name
  def get_weight(self):
    return self.weight
  def get_value(self):
    return self.value
```

```python
class Weapon(Item):
  def __init__(self, name, weight, value, typee, hit_unit):
    super().__init__(name, weight, value)
    self.typee = typee
    self.hit_unit = hit_unit

  def get_hit_unit(self):
    return self.hit_unit
  def get_name(self):
    return self.name
```

```python
class Avatar():
  def __init__(self, posX, posY, name, inventory):
    self.posX = posX
    self.posY = posY
    self.name = name
    self.inventory = inventory

  def say_name(self):
    return f"Hello adventurers my name is {self.name}"

  def moveright(self):
    if(self.posX < 10):
      self.posX += 1
    else:
      print('cannot move right')
    print(f'>>>>>>>>>>>> {self.name}({self.posX}, {self.posY})')

  def moveleft(self):
    if(self.posX > 0):
      self.posX -= 1
    else:
      print('cannot move left')
    print(f'>>>>>>>>>>>> {self.name}({self.posX}, {self.posY})')

  def jump(self):
    if(self.posY < 10):
      self.posY += 0.5
    else:
      print('cannot jump')
    print(f'>>>>>>>>>>>> {self.name}({self.posX}, {self.posY})')

  def godown(self):
    if(self.posY > 0):
      self.posY -= 0.5
    else:
      print('cannot go down')
    print(f'>>>>>>>>>>>> {self.name}({self.posX}, {self.posY})')

  def get_inventory(self):
    return self.inventory

  def get_x(self):
    return self.posX

  def get_y(self):
    return self.posY

  def get_name(self):
    return self.name
```

```python
class Enemy(Avatar):
  def __init__(self, posX, posY, name, inventory, lifepoints, typee):
    super().__init__(posX, posY, name, inventory)
    self.lifepoints = lifepoints
    self.typee = typee

  def update_lifepoints(self, hit_unit):
    self.lifepoints -= hit_unit
    print(f'___________________ {self.name} lifepoints are {self.lifepoints} ___________________')
    return

  def get_lifepoints(self):
    return self.lifepoints

  def drop_inventory(self):
    return self.get_inventory()
  def get_type(self):
    return self.typee
```

```python
class Hero(Avatar):
  def __init__(self, posX, posY, name, inventory, username, lifepoints, energy, main_weapon):
    super().__init__(posX, posY, name, inventory)
    self.username = username
    self.lifepoints = lifepoints
    self.energy = energy
    self.main_weapon = main_weapon

  def hit(self, weapon, enemy):
    hit_unit = weapon.get_hit_unit()
    return enemy.update_lifepoints(hit_unit)

  def full_heal(self):
    self.lifepoints = 1.0

  def check_inventory(self):
    for i in range(len(self.inventory)):
      print(f"Item name: {self.inventory[i].get_name()}, it weighs {self.inventory[i].get_weight()}, and its value {self.inventory[i].get_value()}")

  def special_attack(self, enemy):
    if(self.energy > 0):
      hit_unit = enemy.get_lifepoints() / 2
      enemy.update_lifepoints(hit_unit)
      return enemy
    else:
      pass

  def update_lifepoints(self, hit_unit):
    self.lifepoints -= hit_unit
    print(f'___________________ {self.name} lifepoints are {self.lifepoints} ___________________')
    return


  def update_inventory(self, item):
      self.inventory.append(item)
      self.check_inventory()
      return self.inventory


```

```python
class Boss(Enemy):
  def __init__(self, posX, posY, name, items, lifepoints, typee, game_level, hit_unit, special_attack_unit, special_attack_msg, num_of_minions, weakness, strength, reward):
    super().__init__(posX, posY, name, items, lifepoints, typee)
    self.game_level = game_level
    self.hit_unit = hit_unit
    self.special_attack_unit = special_attack_unit
    self.special_attack_msg = special_attack_msg
    self.num_of_minions = num_of_minions
    self.weakness = weakness
    self.strength = strength
    self.reward = reward

  def normal_hit(self, hero):
    return hero.update_lifepoints(self.hit_unit)

  def special_attack(self, hero):
    print(f"{self.special_attack_msg}")
    return hero.update_lifepoints(self.special_attack_unit)

  def return_reward(self, hero):
    hero.update_inventory(self.reward)
    return hero

```

### ***Creating the game's characters:***

```python
# Creating a hero instance:
weapon1 = Weapon('sword', 2, 50, 'medium', 0.15)
weapon_main = Weapon('retractable_sword', 1, 160, 'good', 0.4)
shield = Item('gold shield', 2, 50)
inventory1 = []
inventory1.append(weapon1)
inventory1.append(weapon_main)
inventory1.append(shield)
hero = Hero(3, 3, 'Gamora', inventory1, 'Baraa', 1.0, 1.0, weapon_main)
```

```python
# Creating enemies:

w3 = Weapon('sword', 1, 40, 'bad', 0.1)
w4 = Weapon('Axe', 2, 60, 'good', 0.25)
w5 = Weapon('other_weapon', 2, 70, 'good', 0.27)
s = Item('golden shield', 2, 100)

inventory2 = []
inventory2.append(w3)
inventory2.append(w4)


inventory3 = []
inventory3.append(w4)
inventory3.append(s)

inventory4 = []
inventory4.append(w3)
inventory4.append(w5)

e1 = Enemy(2, 2, 'Red Skull', inventory2, 1.0, 'medium')
e2 = Enemy(4, 8, 'Loki', inventory3, 1.0, 'strong')
e3 = Enemy(6, 5, 'Nebula', inventory4, 1.0, 'strong')
```

```python
# Creating a boss:

s1 = Item('silver shield', 3, 80)
w6 = Weapon('sword', 2, 60, 'good', 0.25)
w7 = Weapon('hammer', 3, 100, 'very good', 0.6)

inventory5 = []
inventory5.append(s1)
inventory5.append(w6)
inventory5.append(w7)

bossss = Boss(7, 8, 'Thanos', inventory5, 1.0, 'powerful', 4, 0.3, 0.6, 'No resurrection this time', 3, 'blade weapons', 'infinity stones', s1)

```

### ***Game environment:***

```python
# bossss (7, 8)
# e1(2, 2), e2(4, 8), e3(6, 5)
# hero(3, 3)

import random


enemies = []

enemies.append(e1)
enemies.append(e2)
enemies.append(e3)

while(len(enemies) > 0):
  for enemy in enemies:
    while(enemy.get_lifepoints() > 0):

      # Making the enemy move a random step whenever the hero is done with its attack
      e_move = random.randint(1, 4)
      if e_move == 1:
        enemy.moveright()
      elif e_move == 2:
        enemy.moveleft()
      elif e_move == 3:
        enemy.jump()
      else:
        enemy.godown()

      # making the hero choose how to move before attacking the enemy

      move = input(f'Move your hero({hero.get_x()}, {hero.get_y()}):\n - (<) to move left\n - (>) to move right\n - (^) to jump\n - (_) to go down\n')
      if move == '>':
        hero.moveright()
      elif move == '<':
        hero.moveleft()
      elif move == '^':
        hero.jump()
      else:
        hero.godown()


      if((enemy.get_x() <= (hero.get_x() - 2)) or (enemy.get_y() <= (hero.get_y() - 2)) or (enemy.get_y() <= (hero.get_y() + 2)) or (enemy.get_y() <= (hero.get_y() + 2)) ):
          # Allowing the hero to hit only for 3 consecutive times
        for i in range(3):
            # Letting the hero choose the weapon they will use to hit with
          if(enemy.get_lifepoints() > 0):
            hero.check_inventory()
            weapon = int(input(f"choose a weapon to hit {enemy.get_name()}({enemy.get_x()}, {enemy.get_y()}) (enter the weapon number): "))

            hero.hit(hero.get_inventory()[weapon - 1], enemy)

          elif(enemy.get_lifepoints() <= 0):
              print(f"   K.O !\n{enemy.get_name()} is dead")
              print("Now you need to kill:")
              enemies.remove(enemy)
              hero.full_heal()
              for enemy in enemies:
                  print(f"- {enemy.get_name()}")


          elif(hero.get_lifepints() <= 0):
            print(f"GAME OVER ")
            break

        if((enemy.get_x() == hero.get_x()) and (enemy.get_y() == hero.get_y())):
          if(enemy.get_type() == 'strong'):
            hero.update_lifepoints(0.25)
          elif(enemy.get_type() == 'medium'):
            hero.update_lifepoints(0.15)
          else:
            hero.update_lifepoints(0.1)

  if(len(enemies) <= 0):
    print(f"GAME OVER\n  we have a winner \n{hero.say_name()}")
    break







```
